# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def max_min(list_n):
  min=list_n[0]   #assigning a sample min value
  max=list_n[0]   #assigning a sample max value
  for item in list_n:
    if item<=min:
      min=item
    if item>=max:
      max=item
  return min,max

n_list=[2,34,56,78,4,1,82]
range=print(("The minimum and maximum number is"),max_min(n_list))

```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean(n):
  sum=0
  for no in n:
    sum += no
  mean_v=sum/len(n)
  return mean_v

list_n=[2,36,45,9,15]
mean_list=print("Mean of the list is",mean(list_n))
```
