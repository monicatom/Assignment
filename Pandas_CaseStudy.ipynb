```python
mean_revenue=md['Revenue (Millions)'].mean()
md['Revenue (Millions)'].fillna(mean_revenue,inplace=True)

mean_metascore=md['Metascore'].mean()
md['Metascore'].fillna(mean_metascore,inplace=True)
```

17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6  Rating < 7.5: Average
c) Rating  7.5: Excellent

```python
bins=[0,6,7.5,10]
labels=['Poor','Average','Execellent']
md['Categorize_Rating']=pd.cut(md['Rating'],bins=bins,labels=labels,right=True)
md
```
