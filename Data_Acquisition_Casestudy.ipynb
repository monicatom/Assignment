```python
import numpy as np
import pandas as pd
import requests
```

Step 1: Load SpaceX Launch Data from API

 Extract relevant columns: name, date_utc, success, details, rocket



```python
url1='https://api.spacexdata.com/v4/launches'
response=requests.get(url1)
data=response.json()
launches=pd.DataFrame(data)
launches=launches[['name', 'date_utc', 'success', 'details', 'rocket']]
launches
```

Convert date_utc to datetime and extract the year

```python
launches['date_utc'] = pd.to_datetime(launches['date_utc'])
launches['year'] = launches['date_utc'].dt.year
launches
```

Step 2: Load Rocket Metadata
● Use https://api.spacexdata.com/v4/rockets
● Extract id, name, type, active, and stages


```python
url2='https://api.spacexdata.com/v4/rockets'
response=requests.get(url2)
data=response.json()
rocket=pd.DataFrame(data)
rocket=rocket[['id', 'name', 'type', 'active', 'stages']]
rocket
```

Step 3: Merge Launch and Rocket Data

● Join the two DataFrames on rocket ID using


```python
m_data=pd.merge(launches,rocket,how='left',left_on='rocket',right_on='id')
m_data
```

Step 4: Add Simulated Country Information


● Add a new column country and randomly assign one of these values:
['USA', 'Russia', 'India', 'China', 'France']


```python
countries = ['USA', 'Russia', 'India', 'China', 'France']
m_data['country'] = np.random.choice(countries, size=len(m_data))
m_data
```

Step 5: Store Merged Data in SQLite3


● Use sqlite3 to create a connection and save the merged DataFrame as a table named
launches
● Table should contain all merged columns including country

```python
import sqlite3
```

```python
conn=sqlite3.connect('launches.db')
```

```python
m_data.to_sql('launches',conn,index=False)
```

```python
df_from_db=pd.read_sql('SELECT *FROM launches LIMIT 5',conn)
df_from_db
```

Step 6: Run SQL Queries on the Data to analyze
1. Launches by Country
2. Which year had the highest number of launches?
3. Top 5 Missions by Launch Count

```python
cursor=conn.cursor()
cursor.execute(""" SELECT country, COUNT(*) AS launch_count
FROM launches
GROUP BY country
ORDER BY launch_count DESC
""")

rows=(cursor.fetchall())
for row in rows:
  print(row)
```

```python
cursor.execute("""
SELECT year, COUNT(*) AS launch_count
FROM launches
GROUP BY year
ORDER BY launch_count DESC
LIMIT 1
""")

rows=(cursor.fetchall())
for row in rows:
  print(row)
```

```python
cursor.execute("""
SELECT name_x, COUNT(*) AS launch_count
FROM launches
GROUP BY name_x
ORDER BY launch_count DESC
LIMIT 5
""")
rows=(cursor.fetchall())
for row in rows:
  print(row)
```

```python
cursor.execute("""
SELECT name_y, COUNT(*) AS launch_count
FROM launches
GROUP BY name_y
ORDER BY launch_count DESC
LIMIT 5
""")
rows=(cursor.fetchall())
for row in rows:
  print(row)
```
