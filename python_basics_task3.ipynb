1. Write a Python program which accepts the radius of a circle from the user
and compute the area.
Sample input: Enter the radius: 3
Sample output: Area of Circle is 28.59999999999

```python
radius=float(input("Enter radius:"))

area=3.14*radius*radius
print(f"Area of a circle: {area}")
```

2. Write a Python Program to accept the details of a student like name, roll
number and mark and display it.
Sample input: Enter the name: Anisha
Enter the roll number: 21
Enter the mark: 78
Sample output: Name: Anisha
Roll No: 21
Mark: 78

```python
class std:
  def __init__(self,name,roll_no,mark):
    self.name=name
    self.roll_no=roll_no
    self.mark=mark

  def display(self):
    print(f"Name:{self.name}")
    print(f"Roll No:{self.roll_no}")
    print(f"Mark:{self.mark}")

name=input("Enter name:")
roll_no=int(input("Enter roll no:"))
mark=int(input("Enter mark:"))

std_details=std(name,roll_no,mark)
std_details.display()
```

4. Given a range of first 10 numbers, write a Python program to iterate from
start number to the end number and print the sum of the current number
and previous number.
Sample input: 1â€¦.10

```python
start=0
end=10
previous=0
for current in range(start,end+1):
  sum_current_no =current+previous
  print(f"Current no:{current}, Previous no:{previous}, sum of current number: {sum_current_no}")
  previous=current
```

6. Write a Python program to check whether a number is prime or not.
Sample input: Enter the number: 3
Sample output: 3 is a prime number

```python
num=int(input("Enter a number:"))
if num<=1:
  print(f"{num} is not a prime.")
else:
  for i in range(2,num):
    if num%i==0:
      print(f"{num} is not a prime number.")
    else:
      print(f"{num} is a prime.")
```

7. Write a Python program to reverse a list using for loop.
Sample input: 10,40,30,70
Sample output: 70,30,40,10


```python
list3=[10,40,30,70]
reverse_list=[]

for item in range(len(list3)-1,-1,-1):
  reverse_list.append(list3[item])
print(reverse_list)
```

9. Write a Python program to find the maximum of three numbers
Sample input: 34,12,7
Sample output: 34

```python
a=int(input("Enter a number:"))
b=int(input("Enter a number:"))
c=int(input("Enter a number:"))

if a>=b and a>=c:
  print(f"{a} is the maximum among three numbers")
elif b>=a and b>=c:
  print(f"{b} is the maximum among three numbers")
else:
  print(f"{c} is the maximum among three numbers")
```

10. Write a Python program to construct the following pattern, using a
nested for loop.
*
**
***
****
*****
****
***
**
*

```python
row=int(input("No of row needed:"))

for i in range(1,row+1):
  for j in range(i):
    print("*", end="")
  print()
for i in range(row-1,0,-1):
  for j in range(i):
    print("*",end="")
  print()
```
