2. Create a Python program to calculate the factorial of a given number using a function.

```python
def factorial(n):
  fact=1
  for i in range(1,n+1):
    fact =fact*i
  return fact
num=int(input("Enter the number:"))
print("Factorial is ",factorial(num))
```

3. Develop a Python program that checks whether a given word or phrase is a palindrome.

```python
def palindrome(pharse):
  pharse=pharse.lower()
  return pharse==pharse[::-1]

text=input("Enter a text:")
if palindrome(text):
  print("The text is palindrome.")
else:
  print("Not a palindrome.")
```

4. Create a Python program that takes a sentence as input and counts the number of vowels and consonants.

```python
text = input("Enter a sentence: ")

vowels = 0
consonants = 0

for char in text.lower():
    if char.isalpha():
        if char in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Number of vowels:", vowels)
print("Number of consonants:", consonants)

```

5. Develop a Python program that takes two lists as input and returns a new
list containing the common elements between them.

```python
list_a=input("Enter first list:").split()
list_b=input("Enter second list:").split()

new_list=[]

for item in list_a:
  if item in list_b:
    new_list.append(item)

print("Common lists in both list :",new_list)


```

6. Write a Python program that takes a string as input and counts the
frequency of each character in the string

```python
string=input("Enter a string:")

no_occurs={}

for char in string:
  if char in no_occurs:
    no_occurs[char] +=1
  else:
    no_occurs[char] =1
print(f"Number of time letter occurs: {no_occurs}")
```

7.Develop a Python program that checks if a given year is a leap year

```python
def leap_year(year):
  if year%4==0:
    print("The year is leap year.")
  else:
    print("Not a leap year.")
choose_year=int(input("Enter the year"))
leap_year(choose_year)
```

8. Write a Python program that checks if a given number, is a prime number

```python
num=int(input("Enter the number:"))

if num<=1:
  print(f"{num} is not a prime number.")
else:
  for i in range(2,num<=1):
    if num%i==0:
      print(f"{num} is not a prime.")
      break
  else:
      print(f"{num} is a prime.")
```
